mutable VS immutable

```python
# Mutable objects:
# list, dict, set, byte array
# Immutable objects:
# int, float, complex, string, tuple...

# mutable example
samplelist = ['U','C','I']
samplelist[2] = 'B'
print('after change', samplelist)

# immutable example
samplestring = 'UCI'
samplestring[2] = 'B'
# gonna give an error, can't change
print('after change', samplestring)
```

LIST

```python
# Lists are used to store multiple items in a single variable.
# List literals are written within square brackets [ ]. Lists work similarly to strings
# use the len() function and square brackets [ ] to access data, with the first element at index 0. 
colors = ['red', 'blue', 'green']
print(colors[0])    ## red
print(colors[2])    ## green
print(len(colors))  ## 3
```

```python
# Sharing/Copying
# refer list object
x = ['a','b']
y = x
# Both return boolean
# IS. asks whether two references/binding are to the same object
# == second asks only whether the two objects store the same values.
if y==x:
    print('yes')
if y is x:
    print('yes')
# Python creates one list object, and binds it to first to x and then to y (the
# assignment y = x just COPIES into name y the same reference that is in name x,
# making y and x refer to the same object). Here, x is y is True (and x == y is
# therefore True): the two different names x and y refer to the same list object.
```

```python
# more example
examplelist = ['a', 'b', 'c']
examplelist.append('d')        
print('append d',examplelist)
examplelist.insert(0, 'e')        ## insert elem at index 0
print('insert e',examplelist)
examplelist.extend(['f', 'g'])  ## add list of elems at end
print('extend f g',examplelist)  

print(examplelist.index('a') )   ## 1

examplelist.remove('a')         ## search and remove that element
print('remove a', examplelist)
examplelist.pop(1)                  ## removes b
print ('pop index 1 element',examplelist)
```

Slicing

```python
# specify where to start the slicing, and where to end. 
mylist = ['a', 'b', 'c', 'd', 'e', 'f']

# positive -> start from left to right, negative -> right to left
# when 2 argument [start:stop(exclusive)]

print(mylist[1 : -1]) 
# when 3 argument [start:stop(exclusive):step]
print(mylist[1 : 4 : 2]) 
```

```python
# range function
# range(start,stop(exclusive))
lst = list(range(10,20))
print(lst)
# range(start,stop(exclusive), step)
lst = list(range(10,50,10))
print(lst)
print(lst[1:3])
```

```python
# using slice to replace
mylist[0:2] = 'z'    ## replace ['a', 'b'] with ['z']
print(mylist)     ## ['z', 'c', 'd']
```

```python
# when step is negative 
a = list(range(10))
print('list',a)
# Counting from end to the beginning with steps of -1:
print('example1',a[::-1])
print("example2",a[9:4:-1])
```

```python
# practice what will be print
a = list(range(10))
print(a[:-4:-2])
```

Function Arguments and Parameters 

```python
# Whenever we DEFINE a function , Reads the keyword def and notes that a function is being defined.
# we specify the names of its PARAMETERS in its header (in parentheses, separated by commas).

# defines a function of two parameters: x and y. When we define a function we
# (re)bind the function's name to the function object. 

def f(x,y):
    return x**y

# Whenever we CALL a function we specify the values of its ARGUMENTS (also in
# parentheses, separated by commas). 
# f(2,2)
f(5,2*6)
# Calling  f(5,2*6) calls this function with two arguments: the arguments 5 and
# 2*6 and arguments (5 and 12) are bound to their matching parameters in the
# function header (and then the body of the function is executed, which can refer
# to the parameter names). Therefore, function calls happen in three steps (this
# rule is an example of "Python has rules about how things work"): 


# return :  we use it as a return value of a function

#   (1) evaluate the arguments
#   (2) bind the argument values (objects) to the parameter names
#   (3) execute the body of the function, using the values bound to the parameter
#       names
```

```python
# more practice
# Write a  function that sum all numbers in the given list.
sample_list = [1,2,3,4,5]
```

```python
# solution
def sumlist(numbers):
    total = 0
    for x in numbers:
        total += x
    return total
print(sumlist(sample_list))
```

FILE

```python
# "r": Read the file. This is the default setting.
# "w": Write to the file. Creates the file if it doesn’t exist, and erases it if it does.
# "a": Write to the end of a file. Creates the file if it doesn’t exist.
# read method
x = open('sss.txt', 'r') 
## We can do stuff with the file here ##
for i in x:
    print( i,end='')
#    For some reason, the print statement is adding new line characters that are not present in the original file.
#    end='' syntax help you get rid of that
x.close()  # Close the file
```

```python
# write method
string = ("\nThe lizard turns his head but he does not see anything,\n"
		+ "Oh, nothing but a dry leaf.")  # Remember we need \n to start a new line
a = open("sss.txt", "a") # Switch to a to append
a.write(string)
a.close() 

# check
x = open("sss.txt", mode="r") # Switch to r to read back in
all = ""
for line in x:
	all += line
x.close() 
print(all)
```

```python
# with method
# This program will create a new file named test.txt in the current directory if it does not exist. 
# If it does exist, it is overwritten.
with open("test.txt",'w') as f:
    f.write("this is my first file\n")
    f.write("hello:)\n")
    f.write("bye;)")
```

Exception in File Handling

```python
# If file could not be opened, OSError is raised and if it is not found, FileNotFoundError is raised.
try:
    f=open("sss.txt","r")
    while True:
        line=f.readline() # readline() method returns one line from the file.
        print (line, end='')
except FileNotFoundError:
       print ("File is not found")
else:
       f.close()
```

```python
# solution
count = 0
with open ("sss.txt", "r") as file: # r to read back in
    for line in file:
        count += 1
    
print(count)
```
