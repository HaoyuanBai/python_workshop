1.4 Looping

```python
# while loops through a block of code as long as a specified condition is true
x = 10
while x <= 14:
    print(x)
    x += 1 # equal to x =x+1 

print("done!")
# If you just want to loop until some condition is True or False, but you don’t know when that’ll be, use a while loop.
```

```python
# for loop execute a block of code a number of times
# iterate over a sequence (that is either a list, a tuple, a dictionary, a set, or a string).

fruits = ["apple", "banana", "cherry"]
for x in fruits:
    print('fruit', x)
    
for x in "oak":
    print(x)
    
for i in range(4):  # loop from 0 to 3
    print('example 0', i)
        
# use a for loop to iterate through an iterable object
for i in range(5, 8):  # loop from 5 to 7
    print('example 1',i)
        
for i in range(4, 10, 2):  # loop from 4 to 8, skipping by 2 each time
    print('example 2', i)
    
# If you want to loop a number of times that you know in advance, 
# like 10 times, or the number stored in x times, then use a for loop with range().
```

LIST

```python
# Lists are used to store multiple items in a single variable.
# List literals are written within square brackets [ ]. Lists work similarly to strings
# use the len() function and square brackets [ ] to access data, with the first element at index 0. 
colors = ['red', 'blue', 'green']
print(colors[0])    ## red
print(colors[2])    ## green
print(len(colors))  ## 3
```

```python
# Sharing/Copying
# refer list object
x = ['a','b']
y = x
# Both return boolean
# IS.     asks whether two references/binding are to the same object
# == second asks only whether the two objects store the same values.
if y==x:
    print('yes')
if y is x:
    print('yes')
# Python creates one list object, and binds it to first to x and then to y (the
# assignment y = x just COPIES into name y the same reference that is in name x,
# making y and x refer to the same object). Here, x is y is True (and x == y is
# therefore True): the two different names x and y refer to the same list object.
```

```python
# more example
list = ['apple', 'car', 'blue']
list.append('cat')         ## append elem at end
list.insert(0, 'sss')        ## insert elem at index 0
list.extend(['red', 'yellow'])  ## add list of elems at end
print(list)  
print(list.index('car') )   ## 2

list.remove('apple')         ## search and remove that element
print('remove apple', list)
list.pop(1)                  ## removes and returns 'larry'
print (list)
```

Slicing

```python
# access parts of your data.
# specify where to start the slicing, and where to end. 
list = ['a', 'b', 'c', 'd']
print(list[1 : -1])   ## ['b', 'c']
list[0:2] = 'z'    ## replace ['a', 'b'] with ['z']
print(list)     ## ['z', 'c', 'd']
```

Function Arguments and Parameters 

```python
# Whenever we DEFINE a function , we specify the names of its PARAMETERS in its header (
# in parentheses, separated by commas).
# Whenever we CALL a function we specify the values of its ARGUMENTS (also in
# parentheses, separated by commas). The definition below

def f(x,y):
    return x**y

f(2,2)
f(5,2*6)

# defines a function of two parameters: x and y. When we define a function we
# (re)bind the function's name to the function object. 

# Calling  f(5,2*6) calls this function with two arguments: the arguments 5 and
# 2*6 and arguments (5 and 12) are bound to their matching parameters in the
# function header (and then the body of the function is executed, which can refer
# to the parameter names). Therefore, function calls happen in three steps (this
# rule is an example of "Python has rules about how things work"): 



# return :  we use it as a return value of a function

#   (1) evaluate the arguments
#   (2) bind the argument values (objects) to the parameter names
#   (3) execute the body of the function, using the values bound to the parameter
#       names
```

FILE

```python
# "r": Read the file. This is the default setting.
# "w": Write to the file. Creates the file if it doesn’t exist, and erases it if it does.
# "a": Write to the end of a file. Creates the file if it doesn’t exist.
# read method
x = open('sss.txt', 'r') # Opening the text file in the working directory for reading only.
## We can do stuff with the file here ##
for i in x:
    print(i)
x.close()  # Close the file

# The read method will return the entire contents of the text file. 
```

```python
# write method
string = ("\nThe lizard turns his head but he does not see anything,\n"
		+ "Oh, nothing but a dry leaf.")  # Remember we need \n to start a new line
a = open("sss.txt", "a") # Switch to a to append
a.write(string)
a.close() 

# check
x = open("sss.txt", mode="r") # Switch to r to read back in
all = ""
for line in x:
	all += line
x.close() 
print(all)
```

Exception in File Handling

```python
# If file could not be opened, OSError is raised and if it is not found, FileNotFoundError is raised.
try:
    f=open("sss.txt","r")
    while True:
        line=f.readline() # readline() method returns one line from the file.
        print (line, end='')
except FileNotFoundError:
       print ("File is not found")
else:
       f.close()
```
