Accessing values of dictionaries in the list

```python
a_dict = {'Yellow':1, 'Red': 2, 'White': 5}

a_list = ['Black', 'Purple']
a_list.append(a_dict)
print(a_list)
# accessing dictionary values
dict_access = a_list[2]
dict_access1 = a_list[2]['Red']
dict_access2 = a_list[2]['White']

#printing dictionary values
print(dict_access)
print(dict_access1)
print(dict_access2)
```

Updating values of dictionaries in list

```python
a_dict = {'Yellow':1, 'Red': 2, 'White': 5}
a_list = ['Black', 'Purple']

a_list.append(a_dict)
print('orginal', a_list)
# accessing dictionary values

a_list[2]['Red'] = 'Hibiscus'
a_list[2]['White'] = 'Jasmine'

#printing values
print(a_list)
```

```python
# Appending list in Python dictionary
Details = {"apple": 1, 
           "pear": 2, "orange": []}
Details["orange"] += [20, "fresh"]
print(Details)
```

```python
# if method
Details = {"apple": 1, 
           "pear": 2, "orange": []}
if "orange" in Details:
    Details["orange"].append([20, "fresh"])
    print(Details)
```

```python
# Tuple items are ordered, unchangeable, and allow duplicate values.
# Tuple items are indexed, the first item has index [0], the second item has index [1] etc.

# Ordered
# When we say that tuples are ordered, it means that the items have a defined order, and that order will not change.

# Unchangeable
# Tuples are unchangeable, meaning that we cannot change, add or remove items after the tuple has been created.

# Allow Duplicates
# Since tuples are indexed, they can have items with the same value:
```

Sort

```python
# difference bewtween sort and sorted

# sort is a method defined on arguments that are LIST objects; it returns
#       the value None, but it MUTATES its list argument with its values in some
#       specified order

# sorted is a function defined on arguments that are ITERABLE objects; it
#       returns a LIST of its argument's values, with its values in some specified
#       order.

# So, the sort method can be applied only to lists, but the sorted function can
# be applied to any iterable (str, list, tuple, set, dict, etc.) 
```

```python
mylist = ['z', 'd', 'c']
mylist.sort()
print(mylist)
```

```python
try:
    votes_dict = {'Charlie': 20, 'Able': 10, 'Baker': 20, 'Dog': 15}
    votes_dict.sort()
except:
    print('there something wrong')

# AttributeError: 'dict' object has no attribute 'sort'

# The problem is: the sort METHOD is defined only for LIST objects, not for DICT
# objects.
```

```python
# As we saw, Python executes the sorted function by creating a temporary 
# list from all the values produced by theiterable, then it sorts that temporary list, 
# and finally it returns the sortedlist. Here is one example of how the sorted function processes votes_dict. 
# executing sorted(votes_dict) is the same as executing sorted(votes_dict.keys())
# which returns a sorted list of the dict's keys, which we can then iterate over,
# printing every key in votes_dict and its associated integer.
votes_dict = {'Charlie' : 20, 'Able' :  40, 'Baker' : 20, 'Dog' : 15}
for c in sorted(votes_dict):  # same as: for c in sorted(votes_dict.keys())
    print('Candidate', c, 'received', votes_dict[c], 'votes')

print(sorted(votes_dict))
# Python prints a list, the list returned by sorted: the dictionary's keys in sorted order:
```

```python
votes = [('Charlie', 20), ('Able', 10), ('Baker', 20), ('Dog', 15)]
votes.sort()        # call sort method with prefix argument votes
print(votes)
for c,v in votes:   # unpacking assignment
    print('person', c, 'received', v, 'votes')
# The call votes.sort() uses the sort METHOD to sort the LIST (MUTATE it); then
# the for loop iterates over this newly-ordered list and prints the information
# in the list, in the order it now appears. Then, the entire votes list is printed
# after the loop: we will see that the list has been MUTATED and is now sorted.
# Python prints the information sorted alphabetically, in increasing order.
```

```python
votes = [('Charlie', 20), ('Able', 10), ('Baker' ,20), ('Dog', 15)]
for c,v in sorted(votes):  # call sorted function on votes (which is iterable)
    print('Candidate', c, 'received', v, 'votes')
print(votes)
# Here we never sort/mutate the votes list. Instead we use the sorted FUNCTION,
# which takes an ITERABLE argument (a LIST is iterable) and returns a NEW LIST
# that is sorted (leaving votes unchanged). Then we iterate over that returned
# list to print its information (which is a sorted version of votes). But here,
# when we print the votes list at the end, we see that the votes list remained
# unchanged. Python prints
```

```python
# reverse
votes = [('Charlie', 20), ('Able', 10), ('Baker' ,20), ('Dog', 15)]
print(sorted(votes,reverse=True))
# Python prints the list of 2-tuples sorted alphabetically, but now in decreasing
# (reversed) order:
```
