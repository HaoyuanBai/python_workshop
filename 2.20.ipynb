```python
import nltk
#  splitting a large sample of text into words
# also used in NLP
word_data = "It's true He sits on a tree branch watching his butterfly friends dance in the garden. They are so colorful! Red, yellow, blue, orange, purple, green, just like a rainbow!"
nltk_tokens = nltk.word_tokenize(word_data)
print (nltk_tokens)
```

```python
from nltk.tokenize import WordPunctTokenizer
# It separates the punctuation from the words.
WordPunctTokenizer().tokenize("let's see how to tokenize")
```

```python
# tokenize the sentence
sentence_data = "Colorful, the rainbow is colorful. Garden, they are planting flowers in the garden."
nltk_tokens = nltk.sent_tokenize(sentence_data)
print (nltk_tokens)
```

```python
# create a sample dataframe with duplicate rows
data = {
    'A': ['1', '2', '3', '2', '2'],
    'B': ['11', '2', '2', '11', '2'],
    'C': ['2', '2', '7', '5', '2']
}
df = pd.DataFrame(data)

# print the dataframe
print("The original dataframe:\n")
print(df)

# drop duplicates
df_unique = df.drop_duplicates()
# drop_duplicates() function identifies the duplicates
#drops the duplicate rows and just keeps their first occurrence

print("\nAfter dropping duplicates:\n")
print(df_unique)

# DataFrame.drop_duplicates(subset=None, keep='first', inplace=False, ignore_index=False)

# subsetcolumn label or sequence of labels, optional
# Only consider certain columns for identifying duplicates, by default use all of the columns.
# keep{‘first’, ‘last’, False}, default ‘first’
# Determines which duplicates (if any) to keep. - first : Drop duplicates except for the first occurrence. - last : Drop duplicates except for the last occurrence. - False : Drop all duplicates.
# inplacebool, default False
# Whether to drop duplicates in place or to return a copy.
# ignore_indexbool, default False
# If True, the resulting axis will be labeled 0, 1, …, n - 1.
```
