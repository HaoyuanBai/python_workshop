Accessing values of dictionaries in the list

```python
a_dict = {'Yellow':1, 'Red': 2, 'White': 5}

a_list = ['Black', 'Purple']
a_list.append(a_dict)
print(a_list)
# accessing dictionary values
dict_access = a_list[2]
dict_access1 = a_list[2]['Red']
dict_access2 = a_list[2]['White']

#printing dictionary values
print(dict_access)
print(dict_access1)
print(dict_access2)
```

Updating values of dictionaries in list

```python
a_dict = {'Yellow':1, 'Red': 2, 'White': 5}
a_list = ['Black', 'Purple']

a_list.append(a_dict)
print('orginal', a_list)
# accessing dictionary values

a_list[2]['Red'] = 'Hibiscus'
a_list[2]['White'] = 'Jasmine'

#printing values
print(a_list)
```

```python
# Appending list in Python dictionary
Details = {"apple": 1, 
           "pear": 2, "orange": []}
Details["orange"] += [20, "fresh"]
Details["orange"] = Details["orange"]+ [20, "fresh"]
print(Details)
```

```python
# if method
Details = {"apple": 1, 
           "pear": 2, "orange": []}
if "orange" in Details.key():
    Details["orange"].append([20, "fresh"])
    print(Details)
```

```python
# notice
# Having one element within parentheses is not enough.
# need a trailing comma to indicate that is a tuple.
# Creating a tuple having one element
my_tuple = ("hello",)
print(type(my_tuple))  # <class 'tuple'>
my_tuple = ("hello")
print(type(my_tuple))  # <class 'tuple'>
```

```python
# Accessing tuple elements using indexing
my_tuple = ('p','e','t','e','r')
# Index must be an integer
print(my_tuple[0])  
print(my_tuple[-1]) 
```

```python
# change the item
my_tuple = (4, 2, 3, [6, 5])

# TypeError: 'tuple' object does not support item assignment
try:
    my_tuple[1] = 9
except:
    print('something wrong')
```

```python
# However, item of mutable element can be changed
my_tuple[3][0] = 9    # Output: (4, 2, 3, [9, 5])
print(my_tuple)
```

```python
# Tuples can be reassigned
my_tuple = (1,)
print(my_tuple, type(my_tuple))
# notice when only has 1 value
my_tuple = ('peter')
print(my_tuple, type(my_tuple))
```

Sort

```python
mylist = ['z', 'd', 'c']
mylist.sort()
print(mylist)
```

```python
try:
    votes_dict = {'Charlie': 20, 'Able': 10, 'Baker': 20, 'Dog': 15}
    votes_dict.sort()
except:
    print('there something wrong')

# AttributeError: 'dict' object has no attribute 'sort'

# The problem is: the sort METHOD is defined only for LIST objects, not for DICT
# objects.
```

```python
votes_dict = {'Charlie' : 20, 'Able' :  40, 'Baker' : 20, 'Dog' : 15}
for c, s in sorted(votes_dict.items()):  # same as: for c in sorted(votes_dict.keys())
    print('Candidate', c, 'received', s, 'votes')

print(sorted(votes_dict.items()))
# Python prints a list, the list returned by sorted: the dictionary's keys in sorted order:
```

```python
votes = [('Charlie', 20), ('Able', 10), ('Baker', 20), ('Dog', 15)]
votes.sort()        # call sort method with prefix argument votes
print(votes)
for c,v in votes:   # unpacking assignment
    print('person', c, 'received', v, 'votes')
```

```python
votes = [('Charlie', 20), ('Able', 10), ('Baker' ,20), ('Dog', 15)]
for c,v in sorted(votes):  # call sorted function on votes (which is iterable)
    print('Candidate', c, 'received', v, 'votes')
print(votes)
```

```python
# reverse
votes = [('Charlie', 20), ('Able', 10), ('Baker' ,20), ('Dog', 15)]
print(sorted(votes,reverse=True))
# (reversed) order:
```
